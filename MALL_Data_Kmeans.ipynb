```python
import pandas as pd 
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("MALL_Customers.csv")
df
```

```python
plt.scatter(x = df["Annual Income (k$)"],y = df["Spending Score (1-100)"])
plt.savefig("scatter_plot")
```

```python
sns.scatterplot(x = df["Annual Income (k$)"],y = df["Spending Score (1-100)"])
```

#### Model_Training

```python
x = df[["Annual Income (k$)","Spending Score (1-100)"]]
std_scaler = StandardScaler()
x = pd.DataFrame(std_scaler.fit_transform(x),columns=x.columns)
x
```

```python
k_means = KMeans(n_clusters=5,random_state=20)
k_means
```

```python
k_means.fit(x)
```

```python
y_means = k_means.predict(x)
y_means
```

```python
plt.figure(figsize=(4,3))
plt.scatter(x = df["Annual Income (k$)"],y = df["Spending Score (1-100)"],c = y_means,cmap = "rainbow")
plt.savefig("cluster")
```

#### Clusters

```python
x = df[["Annual Income (k$)","Spending Score (1-100)"]]
k_means = KMeans(n_clusters=10,random_state=20)
k_means.fit(x)
y_means = k_means.predict(x)
y_means
```

```python
k_values = np.arange(1,20)
wcss_list = []
for k in k_values:
    k_means = KMeans(n_clusters=k,random_state=20)
    k_means.fit(x)
    wcss_list.append(k_means.inertia_)
plt.plot(k_values,wcss_list)
plt.xlabel("K Values(Number of cluster)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.savefig("elbow_method")
```
